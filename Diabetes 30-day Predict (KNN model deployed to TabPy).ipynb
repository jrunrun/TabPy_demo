```python
import pandas as pd
import numpy as np


#load data from CSV into pandas dataframe
filename = '/Users/jcraycraft/Dropbox/DiabetesPredictiveDemoKit/SourceData/HCPyDiabetesClinical_excludeNone2.csv'
dataset = pd.read_csv(filename)
```

```python
#convert pandas dataframe to numpy array
array = dataset.values
#slice out columns 2, 3, and 4 ('SystolicBPNBR','LDLNBR','A1CNBR')
#>=2 and <5
# define X as array[rows, columns]; grab all rows
X = array[:,2:5]
# define Y as the answer, the flag of Y/N for 30-day Readmit based off of the outcomes we have recorded 
# for each of our patiens
Y = array[:,6]
```

```python
# print array X
X
```

```python
# print array Y
Y
```

```python
from sklearn.model_selection import train_test_split

#from sklearn.cross_validation import train_test_split

from sklearn.neighbors import KNeighborsClassifier


#split out train/test/split at 80/20
validation_size = 0.20
#define seed for randomness; in other words, we want to control the random if we do this 
#several times as we move through feature selection in order to accurately compare results
#same random number each time will ensure we have same split of data to ensure apples to apples
seed = 7
#randomly selects test/compare, rinse repeate for all the data
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=validation_size, random_state=seed)
# Make predictions on validation dataset
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
predictions = knn.predict(X_validation)
predictions_proba = knn.predict_proba(X_validation)
```

```python
# The predictive function called Diabetes30DayReadmit that will return Y/N for 30-day Readmit Risk
def Diabetes30DayReadmit(bp, ldl, a1c):
    X = np.column_stack([bp, ldl, a1c])
    return knn.predict(X).tolist()
```
